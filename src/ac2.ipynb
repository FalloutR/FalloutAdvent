```python
#Advent of Code problem 2
#implemented below, will use on next problem :)
#plot the path of the sub? use better "structure"?

import numpy as np
import matplotlib.pyplot as plt

f_in = open("../data/ac2.in")

input_list = f_in.readlines()

x = 0 # x position of sub
y = 0 # y position of sub

for ea in input_list:
    instruction = ea.split() #split the input so we have [direction,intensity]
    direction = instruction[0]
    intensity = int(instruction[1]) #parse intensity to int
    
    #print("Direction: "+direction[0]) #following if parses direction into x and y movement of sub
    if direction[0] == "f":
        print("Going forward!")
        x+=intensity
    elif direction[0] == "d":
        print("Going down!")
        y+=intensity
    elif direction[0] == "u":
        print("Going up!")
        y-=intensity
    elif direction[0] == "b":
        print("Going backwards!") #unused
    else:
        print("UNKNOWN DIRECTION!!!") #unused
    
    
print("x: "+str(x)+", y: "+str(y))
print("solution: "+str(x*y))
```

```python
#Advent of Code problem 2 part 2

import numpy as np
import matplotlib.pyplot as plt

f_in = open("../data/ac2.in")

input_list = f_in.readlines()

x = 0 # x position of sub
y = 0 # y position of sub

angle = 0 # angle or "aim" of the sub

for ea in input_list:
    instruction = ea.split() #split the input so we have [direction,intensity]
    direction = instruction[0]
    intensity = int(instruction[1]) #parse intensity to int
    
    #print("Direction: "+direction[0]) #following if parses direction into x and y movement of sub
    if direction[0] == "f":
        print("Moving!")
        x+=intensity
        y+=angle*intensity
    elif direction[0] == "d":
        print("Adjusting aim down!")
        angle+=intensity
    elif direction[0] == "u":
        print("Adjusting aim up!")
        angle-=intensity
    elif direction[0] == "b":
        print("Going backwards????") #unused
    else:
        print("UNKNOWN DIRECTION!!!") #unused
    
    
print("x: "+str(x)+", y: "+str(y))
print("solution: "+str(x*y))
```
